```python
import pandas as pd

# Read the CSV file
df = pd.read_csv('evaluation/all_evaluations.csv')

# Display the dataframe
df.head
```

```python
# Define a function to extract and print TP, FP, FN, TN values for a given model and the highest F1 score
def print_confusion_matrix_for_highest_f1(model_name):
    # Filter the DataFrame for the given model
    model_df = df[df['Model Name'] == model_name]
    
    # Convert the string representations of lists into actual lists
    f1_scores = model_df['F1'].apply(lambda x: eval(x)).iloc[0]
    tp_list = model_df['TP'].apply(lambda x: eval(x)).iloc[0]
    fp_list = model_df['FP'].apply(lambda x: eval(x)).iloc[0]
    fn_list = model_df['FN'].apply(lambda x: eval(x)).iloc[0]
    tn_list = model_df['TN'].apply(lambda x: eval(x)).iloc[0]

    # Find the index of the highest F1 score
    highest_f1_index = f1_scores.index(max(f1_scores))
    
    # Extract the TP, FP, FN, TN values using the index found
    tp = tp_list[highest_f1_index]
    fp = fp_list[highest_f1_index]
    fn = fn_list[highest_f1_index]
    tn = tn_list[highest_f1_index]

    # Print the confusion matrix values
    print(f"Model: {model_name}")
    print(f"Highest F1 score: {f1_scores[highest_f1_index]} at index {highest_f1_index}")
    print(f"TP: {tp}, FN: {fn}, TN: {tn}, FP: {fp}")



```

```python
# Let's use the first model in the dataframe as an example
model_to_search = df['Model Name'][7]

print_confusion_matrix_for_highest_f1(model_to_search)
```
